```python
# Function to preprocess the image
def preprocess_image(image):
    # Enhance edges using adaptive histogram equalization (CLAHE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced_image = clahe.apply(image)
    
    # Apply a bilateral filter to remove noise while keeping edges sharp
    filtered_image = cv2.bilateralFilter(enhanced_image, d=9, sigmaColor=75, sigmaSpace=75)
    
    return enhanced_image, filtered_image
```

```python
epsilon = 1e-7
def dice_coeff(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    intersection = K.sum(y_true_f * y_pred_f)
    dice = (2. * intersection + epsilon) / ( K.sum(y_true_f) + K.sum(y_pred_f) + epsilon)
    return dice
```

```python
def dice_loss(y_true, y_pred):
    dice = dice_coeff(y_true, y_pred)
    return 1 - dice
```

```python
# Load the model, ensuring you include the custom objects in the call
unet = load_model('Midterm_Model_1000_enhanced_images.h5', custom_objects={'dice_coeff': dice_coeff, 'dice_loss': dice_loss})
```

```python
import numpy as np
from scipy.spatial import distance
from skimage.measure import label, regionprops

def extract_key_points(mask, threshold=0.5):
    labeled_mask = label(mask > threshold)
    properties = regionprops(labeled_mask)
    key_points = [prop.centroid for prop in properties]  # Get centroids of labeled regions
    return np.array(key_points)

def calculate_correct_keypoints_percentage(true_masks, pred_masks, threshold=5):
    correct_count = 0
    total_count = 0
    

    # Squeeze the pred_masks to remove the singleton dimension
    pred_masks = np.squeeze(pred_masks)

    
    for true_mask, pred_mask in zip(true_masks, pred_masks):
        true_key_points = extract_key_points(true_mask)
        pred_key_points = extract_key_points(pred_mask)
        
        total_count += len(true_key_points)
        
        for pred_point in pred_key_points:
            if any(distance.euclidean(pred_point, true_point) <= threshold for true_point in true_key_points):
                correct_count += 1

    if total_count == 0:
        return 0  # To avoid division by zero if there are no key points
    return (correct_count / total_count) * 100

# Example usage
threshold_distance = 5  # pixels

percentage_correct = calculate_correct_keypoints_percentage(test_masks_stacked, pred_masks, threshold=threshold_distance)
print(f"Percentage of Correct Key Points: {percentage_correct}%")
```
